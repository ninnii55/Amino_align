## (Capital) GAmma Function (Decay constaant) Calculation

$\Gamma$ = -D$q^{2}$

$q$ = wave/scatter vector
q = $(4*(\pi)n_{O}/\lambda)*sin(\theta/2)$

$n_{O}$ = solvent refractive index
$\theta$ = Scatter angle
$\lambda$ = Wavelength of light used

Note: Instument used is Nano Zetasizer NS range, wavelngth used indicative of Mei scattering
$\theta$ = 173 degrees
$\lambda$ = 633 nm


## $\Gamma$ function with respect to gelation, homo and heterodyne scattering influences the decay of signal. Therefore, Gamma changes to one of the following :

## $\Gamma$ = $exp$[-2D$q^{2}\tau$] (homodyne)
## $\Gamma$ = $exp$[-D$q^{2}\tau$] (heterodyne)
## $D_t=\ \frac{k_BT}{3\pi\eta R_H}\left(ln\left(\frac{L}{d}\right)+0.32\right)$
$d = diameter of cylinder$
$\left(\frac{L}{d}\right)=aspect\ ratio(for\ nano\ tubes$
$L=Lenght\ of\ cylender$

```python
def alnSiteCompositionDF(aln, characters="ACDEFGHIKLMNPQRSTVWY"):
  alnRows = aln.get_alignment_length()
  compDict = {char:[0]*alnRows for char in characters}
  for record in aln:
    header = record.id
    seq = record.seq
    for aaPos in range(len(seq)):
      aa = seq[aaPos]
      if aa in characters:
        compDict[aa][aaPos] += 1
  return pd.DataFrame.from_dict(compDict)
```

```python
RT_alignmentSiteCompDF = alnSiteCompositionDF(RT_align)
INT_alignmentSiteCompDF = alnSiteCompositionDF(INT_align)
PRO_alignmentSiteCompDF = alnSiteCompositionDF(PRO_align)
```
